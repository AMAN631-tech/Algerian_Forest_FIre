# ***Algerian Forest Fire Dataset – Introduction***
This dataset records daily weather conditions and fire occurrences in forests of Algeria, specifically in the summer of 2012. It includes environmental variables like temperature, humidity, wind speed, and fire indices, along with whether a fire occurred on that day or not. This dataset is useful for analyzing and predicting forest fire risks.
### Fire Indices ###
Fire indices are special calculated values that estimate the risk, behavior, and intensity of a forest fire based on weather and environmental conditions.

### # Column Descriptions #####

| **Column Name** | **Description**                      | **Explanation**                                            |
| --------------- | ------------------------------------ | --------------------------------------------------------------- |
| `day`           | Day of the month (1–31)              | Shows which day the data was collected                          |
| `month`         | Month of the year (e.g., 6 for June) | Tells us the month, mainly June in this dataset                 |
| `year`          | Year of observation (e.g., 2012)     | Shows the year of data collection                               |
| `Temperature`   | Daily temperature in Celsius         | How hot it was that day                                         |
| `RH`            | Relative Humidity in %               | How much moisture was in the air                                |
| `Ws`            | Wind speed in km/h                   | How fast the wind was blowing                                   |
| `Rain`          | Rainfall in mm                       | How much rain fell that day                                     |
| `FFMC`          | Fine Fuel Moisture Code              | Indicates dryness of small fuels like grass (higher = drier)    |
| `DMC`           | Duff Moisture Code                   | Indicates dryness of medium-sized forest material (like leaves) |
| `DC`            | Drought Code                         | Long-term dryness of deep soil and heavy fuels                  |
| `ISI`           | Initial Spread Index                 | Measures how fast a fire might spread at the start              |
| `BUI`           | Build-Up Index                       | Combines DMC and DC to estimate total fuel availability         |
| `FWI`           | Fire Weather Index                   | Final value estimating fire intensity                           |
| `Classes`       | Fire status (`fire` or `not fire`)   | Tells whether a fire occurred that day                          |
